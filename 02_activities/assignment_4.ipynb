```python
import geopandas as gpd

# Load the GeoJSON file using a relative path
file_path = "assignments/toronto-beaches-water-quality.geojson"
beaches_data = gpd.read_file(file_path)

# Display the columns in the DataFrame
print(beaches_data.columns)

# Display the first few rows to understand the data
print(beaches_data.head())


```

```python
# Display the columns in the DataFrame
print(beaches_data.columns)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert the date column to datetime format
beaches_data['collectionDate'] = pd.to_datetime(beaches_data['collectionDate'])

# Plot E. Coli levels over time for each beach
plt.figure(figsize=(12, 8))
sns.lineplot(data=beaches_data, x='collectionDate', y='eColi', hue='beachName')
plt.title('E. Coli Levels Over Time at Toronto Beaches')
plt.xlabel('Date')
plt.ylabel('E. Coli Count (CFU/100mL)')
plt.legend(title='Beach')
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
5
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
fig,(ax1, ax2) = plt.subplots(ncols=2,
                        nrows=1,
                        figsize=(7, 3))
6
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                        nrows=1,
                        figsize=(7, 3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
fig.show()
7
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4))
for label, ax in someaxes.items():
 ax.text(0.5, 0.5, f'This is {label!r}',
        fontsize=14,
        ha = 'center',
        transform=ax.transAxes)
9
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(9,3))

font1 = {'family': 'serif', 'size': 14}

ax1.scatter(x1, y1, marker='o', color='red')
ax2.bar(x2, y2, color='darkgreen')

ax1.set_title('Scatterplot!', fontdict=font1)
ax2.set_title('Barplot!', fontdict=font1)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                   figsize=(7, 4))

for label, ax in someaxes.items():
    ax.text(0.5, 0.5, f'This is {label!r}',
            fontsize=14,
            ha='center',
            transform=ax.transAxes)

10
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
11
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4)) 
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
plt.show()
13
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4),
                                layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
plt.show()
16
```

```python
# first make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])
# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))
# now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1,
color = "indigo")
ax.plot(x, y2,
color = "red")
19
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
21
```

```python
ax.errorbar(x, #our x values
        y2, #our y values
        yerr = y2_sd,
        fmt = "none")
22
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2,
        color = "red")
ax.errorbar(x,
        y2,
        yerr = y2_sd,
        fmt = "none",
        ecolor= "indigo",
        elinewidth= 4,
        capsize = 6,
        capthick= 4 )
23
```
